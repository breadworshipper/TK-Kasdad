```python
# import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df_result_preprocessing = pd.read_csv('preprocessing_result.csv')
df_result_preprocessing.head(99)
```

```python
df_result_model = pd.read_csv('model_result.csv')
df_result_model.head(99)
```

```python
plt.figure(figsize=(10, 5))
df_result_model.plot(x='model', y='r_squared', kind='bar')
plt.xlabel('Model')
plt.ylabel('R-squared')
plt.title('Model R-squared Results')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df_result_model.plot(x='model', y='rmse', kind='bar')
plt.xlabel('Model')
plt.ylabel('RMSE')
plt.title('Model RMSE Results')
plt.show()
```
